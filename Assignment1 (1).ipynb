Age x Income Comparison

```python
under50 = df[(df['income'] == '<=50K')]
over50 = df[(df['income'] == '>50K')]
```

```python
under50.describe()
```

```python
over50.describe()
```

The data shows that the average age of people who make under \$50K is 8 years younger than the average age of those who make over \$50K (36 and 44, respectively).

Histograms Comparing Hours Per Week and Income

```python
under50.plot(kind='hist', x='income', y='hours-per-week', title="<=50K Hours Per Week")
```

```python
over50.plot(kind='hist', x='income', y='hours-per-week', title="<=50K Hours Per Week")
```

Analysis of the histograms reveals that people who make greater than \$50k generally work longer hours than people who make under \$50K. There is a significant spike in frequency among the 40-70 hours per week chart in the greater than \$50k hours per week chart that is not present in the less than \$50k hours per week chart.
